```python
# make a dataset with each author1's name and the number of weeks their book was on the bestseller list

df_authors = df.groupby('author1').size().reset_index(name='weeks_on_list')
```

```python
# make a histogram of the number of weeks each author has been on the bestseller list if they have weeks on the list greater than 100

import matplotlib.pyplot as plt

# Filter the DataFrame
filtered_df = df_authors[df_authors['weeks_on_list'] > 52]

# Plot the histogram
filtered_df['weeks_on_list'].plot.hist(bins=20)
plt.xlabel('Weeks on List')
plt.ylabel('Frequency')
plt.title('Histogram of Weeks on List (More than 100 Weeks)')
plt.show()

```

```python
# extract country from place column

df['country'] = df['place'].str.extract(r'(\w+)$')

```

```python
df.head()
```

```python
# show unique values in the country column

df['country'].unique()
```

```python
# show values where country is 'Federation'

df[df['country'] == 'Africa']   
```

```python
# rename countries

df['country'] = df['country'].replace('States', 'United States')
df['country'] = df['country'].replace('Missouri', 'United States')
df['country'] = df['country'].replace('Massachusetts', 'United States')
df['country'] = df['country'].replace('Federation', 'Russia')
df['country'] = df['country'].replace('Carolina', 'United States')
df['country'] = df['country'].replace('Republic', 'Dominican Republic')
df['country'] = df['country'].replace('Alabama', 'United States')
df['country'] = df['country'].replace('Michigan', 'United States')
df['country'] = df['country'].replace('Kingdom', 'United Kingdom')
df['country'] = df['country'].replace('Lanka', 'Sri Lanka')
df['country'] = df['country'].replace('Kentucky', 'United States')
df['country'] = df['country'].replace('Zealand', 'New Zealand')
df['country'] = df['country'].replace('Ohio', 'United States')
df['country'] = df['country'].replace('IA', 'United States')
df['country'] = df['country'].replace('of', 'South Korea')
df['country'] = df['country'].replace('Africa', 'South Africa')
df['country'] = df['country'].replace('Kong', 'Hong Kong')
df['country'] = df['country'].replace('Texas', 'United States')
```

```python
# create a columns for US and non-US authors

df['us_author'] = df['country'].apply(lambda x: 'Yes' if x == 'United States' else 'No')
```

```python
# create a chart of us and non-us authors over time

df_us_authors = df.groupby(['year', 'us_author']).size().reset_index(name='count')

df_us_authors.head()

df_us_authors_pivot = df_us_authors.pivot(index='year', columns='us_author', values='count').reset_index()

df_us_authors_pivot.head()

df_us_authors_pivot.plot(x='year', y=['Yes', 'No'], kind='bar', stacked=True)

plt.xlabel('Year')
plt.ylabel('Number of Authors')
```

```python
# make the same chart by all countries

df_countries = df.groupby(['year', 'country']).size().reset_index(name='count')

df_countries_pivot = df_countries.pivot(index='year', columns='country', values='count').reset_index()

df_countries_pivot.plot(x='year', y=['United States', 'Russia', 'United Kingdom', 'South Korea', 'South Africa', 'Hong Kong', 'Sri Lanka', 'New Zealand', 'Dominican Republic'], kind='bar', stacked=True)

plt.xlabel('Year')
plt.ylabel('Number of Authors')

plt.show()
```
