```python
import pandas as pd
```

#### Import the file with books and authors

```python
df_books = pd.read_csv('goodreads-books.csv') 
df_books.head()
```

```python
# stripping line break character from good_reads_book 

df_books['good_reads_book']= df_books['good_reads_book'].str.strip('\n')
# df_books.drop_duplicates(inplace=True)
```

```python
# getting ratings and no. of reveiewers into two separate columns

# Splitting the 'book_rating' column into two parts: 'avg rating' and 'ratings'
df_books[['avg_rating', 'no_of_ratings']] = df_books['book_rating'].str.extract(r'(\d+\.\d+) avg rating — ([\d,]+) ratings')

# Converting 'avg_rating' to float
df_books['avg_rating'] = df_books['avg_rating'].astype(float)

# Converting 'ratings' to integer (after removing commas and handling NaN)
df_books['no_of_ratings'] = df_books['no_of_ratings'].str.replace(',', '')
df_books['no_of_ratings'] = pd.to_numeric(df_books['no_of_ratings'], errors='coerce')

```

#### Merge this with the original NYT bestsellers list to get a master file and see missing values

```python
# import the NYT raw data 

df_nyt = pd.read_csv('nyt_raw.csv')
df_nyt.shape
```

```python
# new_df = pd.merge(A_df, B_df,  how='left', left_on=['A_c1','c2'], right_on = ['B_c1','c2'])

df_merged = df_nyt.merge(df_books, left_on=['book','author1'], right_on=['nyt_name','author'], how='left')
df_merged.info()
```

```python
# extract year from the date

df_merged['date']= pd.to_datetime(df_merged['date'])
df_merged['year'] = df_merged['date'].dt.year
```

```python
df_merged.drop(columns= 'Unnamed: 0', inplace=True)
df_merged.head()
```

#### Get authors information

```python
# get df from author scrape
df_authors = pd.read_csv('aithors_info.csv')
```

```python
# Remove line breaks and the word 'in' from the 'author_place' column
df_authors['author_place'] = df_authors['author_place'].str.replace('\n', '').str.strip()
df_authors['author_place'] = df_authors['author_place'].str.replace('in ', '', regex=False)
df_authors.head()

df_authors.drop(columns=['Unnamed: 0', 'author_genres'], inplace=True)

```

```python
df_authors.drop_duplicates(inplace=True)

df_filtered = df_authors[~(df_authors['author_place'] == 'Not available')]

df_filtered.to_csv('author-address.csv')

```

```python
# import the author addresses with lats and longs (from geolocating.ipynb)

df_addresses = pd.read_csv('gr_author_address.csv')

df_merged_final = df_merged.merge(df_addresses, left_on='author_link', right_on='gr_link', how='left')
```

```python
df_merged_final.head()
```

```python
df_merged_final.drop(columns=['book_rating', 'gr_link', 'author_y'], inplace=True)
```

```python
df_merged_final = df_merged_final.rename(columns={'author_x': 'authors'})
df_merged_final.drop(columns=['Unnamed: 0.1',  'Unnamed: 0'], inplace=True)
df_merged_final.head()
```

```python
# save this clean data set as a CSV

df_merged_final.to_csv('df_nyt_merged_cleaned.csv')
```
